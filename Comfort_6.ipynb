### ***Task 1 — PRODUCT VARIABLES***

```python
product_name = 'KNB Jug'
product_price = 19000.00
```

### ***Task 2 — COLLECT CUSTOMER FEEDBACK***

```python
Ratings = []

rating_1 = 4
rating_2 = 3
rating_3 = 3
rating_4 = 4
rating_5 = 5
```

```python
Ratings.append(rating_1)
Ratings.append(rating_2)
Ratings.append(rating_3)
Ratings.append(rating_4)
Ratings.append(rating_5)
print(Ratings)
```

### ***Task 3  — MODIFY THE LIST***

```python
rating_6 = 5
Ratings.append(rating_6)
```

```python
Ratings.pop(0)
```

```python
Ratings.insert(2, 4)
Ratings
```

### ***Task 4 — ANALYZE THE DATA***

```python
Highest_rating = max(Ratings)
Highest_rating
```

```python
Lowest_rating = min(Ratings)
Lowest_rating
```

```python
Average_rating = sum(Ratings)/len(Ratings)
Average_rating
```

```python
Number_of_5_star_ratings = len(Ratings)
Number_of_5_star_ratings
```

```python
Sorted_Ratings = Ratings.sort()
print(Ratings)
```

### ***Task 5 — SUMMARY***

```python
print("-------------Customer Feedback Summary-------------")

print(f"\nProduct: {product_name}")
print(f"Price: {product_price} ")
print(f"Final Ratings: {Sorted_Ratings}")
print(f"Highest Rating: {Highest_rating}")
print(f"Lowest Rating: {Lowest_rating}")
print(f"Average Rating: {Average_rating}")
print(f"Number of 5-star ratings: {Number_of_5_star_ratings}")
```
